```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python

def draw_pert_chart(tasks, dependencies):
    # Create a directed graph
    G = nx.DiGraph()

    # Add nodes with their durations
    for task, duration in tasks.items():
        G.add_node(task, duration=duration)

    # Add edges based on dependencies
    for dep in dependencies:
        G.add_edge(dep[0], dep[1])

    # Compute positions for each node using graph layout
    pos = nx.planar_layout(G)  # You can experiment with different layouts like spring_layout

    # Draw the nodes
    nx.draw(G, pos, with_labels=True, node_size=2000, node_color='skyblue')

    # Draw the edge labels (dependencies)
    edge_labels = {(u, v): f'{G.nodes[u]["duration"]}' for u, v in G.edges()}
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

    # Show the plot
    plt.show()
```

```python
# Example tasks and durations
tasks = {
    'Start': 0,
    'A': 5,
    'B': 3,
    'C': 2,
    'D': 4,
    'E': 2,
    'Finish': 0
}

# Dependencies between tasks (A depends on Start, B depends on A, etc.)
dependencies = [
    ('Start', 'A'),
    ('Start', 'B'),
    ('A', 'C'),
    ('B', 'C'),
    ('C', 'D'),
    ('C', 'E'),
    ('D', 'Finish'),
    ('E', 'Finish')
]

draw_pert_chart(tasks, dependencies)
```
